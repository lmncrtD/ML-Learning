```python
my_list = [1, 3, 5, 7, 9]
```

```python
def binary_search(arr, item):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        guess = arr[mid]
        if guess == item:
            return mid
        elif guess > item:
            high = mid - 1
        elif guess < item:
            low = mid + 1
        else:
            return None
```

```python
print(binary_search(my_list, 1))
```
