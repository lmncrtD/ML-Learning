```python
nums = [1,3,5,6]
```

```python
class Solution:
    def searchInsert(self, nums: list[int], target: int) -> int:
        low = 0
        high = len(nums) - 1
        while low <= high:
            mid = (low + high) // 2
            guess = nums[mid]
            if guess == target:
                return mid
            elif guess < target:
                low = mid + 1
            elif guess > target:
                high = mid - 1
        return low # Почему low? Потому что если мы не находим элемента, low перемещается туда, где должен находиться таргет.
            
```

```python
search_index = Solution()
```

```python
search_index.searchInsert(nums, 2)
```
