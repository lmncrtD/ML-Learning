```python
nums = [-1,0,3,5,9,12]
```

```python
class Solution:
    def search(self, nums: list[int], target: int) -> int:
        self.nums = nums
        self.target = target
        low = 0
        high = len(nums) - 1
        while low <= high:
            mid = (low + high) // 2
            guess = nums[mid]
            if guess == target:
                return mid
            elif guess < target:
                low = mid + 1
            elif guess > target:
                high = mid - 1
            else:
                return -1
```

```python
search_target = Solution().search(nums, 3)
```
